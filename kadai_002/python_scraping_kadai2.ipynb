```python
import requests
import pprint
from getpass import getpass

api_key = getpass('APIキーを入力してください：')
```

```python
search_word = input('検索する駅を入力してください：')
```

```python
url = 'https://maps.googleapis.com/maps/api/geocode/json'
response = requests.get(
    url,
    params={
        'key': api_key,
        'address': search_word,
    }
)
```

```python
json_data = response.json()

pprint.pprint(json_data)
```

```python
location = json_data['results'][0]['geometry']['location']
pprint.pprint(location)
pprint.pprint(location['lat'])
```

```python
url_location ='https://maps.googleapis.com/maps/api/place/nearbysearch/json'

response_location = requests.get(
    url_location,
    params={
        'key': api_key,
        'type':'restaurant',
        'language':'ja',
        'location':f'{location["lat"]},{location["lng"]}',
        'radius':500,
    }
)
```

```python
location_data = response_location.json()
pprint.pprint(location_data)
```

```python
for restaurant in location_data['results']:
    restaurant_name = restaurant['name']
    restaurant_rate = restaurant['rating']
    restaurant_address = restaurant['vicinity']
    print(f'{restaurant_name}:{restaurant_rate}:{restaurant_address}')
```
