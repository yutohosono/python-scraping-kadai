```python
import requests
from bs4 import BeautifulSoup

# タイトルを取得する対象のURL
url = 'https://news.yahoo.co.jp/articles/22b6487d112a88b9c18ee1a09fcce15c1665ce80'

# GETリクエストを送信し、HTMLを取得する
response = requests.get(url)
# HTMLを解析する
soup = BeautifulSoup(response.text, 'html.parser')

# ページのタイトル要素を直接取得する
# タイトルの要素を取得する
# CSSセレクタを使用して、タイトルを取得する
article_element = soup.select_one('#uamods > .article_body')
p_element = article_element.find_all('p')


for p in p_element:

    print(p.get_text(strip=True))



```

上記のコードを実行すると、ページのHTMLソースコードが出力されます。この出力結果を注意深く見て、記事のタイトルがどのHTMLタグ（例: `<h1>`, `<p>`, `<span>` など）や属性（`id`, `class` など）の中に含まれているかを探してください。出力されたHTMLが長すぎる場合は、ブラウザの開発者ツールで確認することをお勧めします。
